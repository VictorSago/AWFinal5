```python
import math

lang_set = set()
for s in df2022["LanguageHaveWorkedWith"]:
    if type(s) == str:
        lang_set.update(s.split(";"))
    elif type(s) == float:
        if math.isnan(s):
            lang_set.add("N/A")

print(lang_set)
```

```python
from collections import Counter

lang_want_counter2022 = Counter()
for s in df2022["LanguageWantToWorkWith"]:
    if type(s) == str:
        lang_want_counter2022.update(s.split(";"))
    elif type(s) == float:
        if math.isnan(s):
            lang_want_counter2022.update(["N/A"])

lang_want_counter2022.most_common()
```
