```python
df2022_prep = df2022.drop(["ResponseId", "Currency", "CompTotal", "CompFreq", "Accessibility", "MentalHealth",
                          "TBranch", "ICorPM", "ConvertedCompYearly"], axis=1)
df2022_prep = df2022_prep.drop([f"Knowledge_{n}" for n in range(1, 8)], axis=1)
df2022_prep = df2022_prep.drop([f"Frequency_{n}" for n in range(1, 4)], axis=1)
df2022_prep = df2022_prep.drop([f"TrueFalse_{n}" for n in range(1, 4)], axis=1)
df2022_prep = df2022_prep.drop(["TimeSearching", "TimeAnswering", "Onboarding", "ProfessionalTech", 
                                "SurveyLength", "SurveyEase"], axis=1)
df2022_prep = df2022_prep.drop(["PurchaseInfluence"], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

mb_encode = LabelEncoder()
df2022_prep["MainBranch"] = df2022_prep["MainBranch"].astype('category')

#Integer encoding the 'type' column
df2022_prep['MainBranch'] = mb_encode.fit_transform(df2022_prep.MainBranch)
df2022_prep
```

```python
rw_encode = LabelEncoder()
df2022_prep["RemoteWork"] = df2022_prep["RemoteWork"].astype("category")
df2022_prep["RemoteWork"] = rw_encode.fit_transform(df2022_prep.RemoteWork)
df2022_prep
```

```python
edlev_encode = LabelEncoder()
df2022_prep["EdLevel"] = df2022_prep["EdLevel"].astype("category")
df2022_prep["EdLevel"] = edlev_encode.fit_transform(df2022_prep.EdLevel)

bchain_encode = LabelEncoder()
df2022_prep["Blockchain"] = df2022_prep["Blockchain"].astype("category")
df2022_prep["Blockchain"] = bchain_encode.fit_transform(df2022_prep.Blockchain)
df2022_prep
```

```python
gender_encode = LabelEncoder()
df2022_prep["Gender"] = df2022_prep["Gender"].astype("category")
df2022_prep["Gender"] = gender_encode.fit_transform(df2022_prep.Gender)

trans_encode = LabelEncoder()
df2022_prep["Trans"] = df2022_prep["Trans"].astype("category")
df2022_prep["Trans"] = trans_encode.fit_transform(df2022_prep.Trans)

df2022_prep
```

```python
sexuality_encode = LabelEncoder()
df2022_prep["Sexuality"] = df2022_prep["Sexuality"].astype("category")
df2022_prep["Sexuality"] = gender_encode.fit_transform(df2022_prep.Sexuality)

ethnicity_encode = LabelEncoder()
df2022_prep["Ethnicity"] = df2022_prep["Ethnicity"].astype("category")
df2022_prep["Ethnicity"] = trans_encode.fit_transform(df2022_prep.Ethnicity)

df2022_prep
```
