```python
df2022 = pd.read_csv("../rawdata/2022/survey_results_public.csv", dtype=str)

df2022 = df2022[['ResponseId', 'EdLevel', 'YearsCode', 'Country', 'OrgSize', 'Age', 'Gender']]
```

```python
df2022
```

```python
for i in df2022.index:
    if str(df2022.at[i, "Gender"]) == "nan":
        df2022.at[i, "Gender"] = "Unknown"
    elif df2022.at[i, "Gender"] not in ["Man", "Woman"]:
        df2022.at[i, "Gender"] = "Other"

df2022
```

```python
list(df2022["Gender"])
```

```python
from collections import Counter

g_cnt = Counter()

g_cnt.update(list(df2022["Gender"]))

g_cnt.most_common()
```

```python
ed_cnt = Counter()

ed_cnt.update(list(df2022["EdLevel"]))

ed_cnt.most_common()
```

```python
for i in df2022.index:
    if str(df2022.at[i, "EdLevel"]) == "nan":
        df2022.at[i, "EdLevel"] = "Unknown"

df2022
```

```python
list(df2022["EdLevel"])
```

```python
ed_cnt = Counter()

ed_cnt.update(list(df2022["EdLevel"]))

ed_cnt.most_common()
```

```python
for i in df2022.index:
    if str(df2022.at[i, "YearsCode"]) == "nan":
        df2022.at[i, "YearsCode"] = "Unknown"
    if str(df2022.at[i, "Country"]) == "nan":
        df2022.at[i, "Country"] = "Unknown"
    if str(df2022.at[i, "Age"]) == "nan":
        df2022.at[i, "Age"] = "Unknown"

df2022
```

```python
org_cnt = Counter()

org_cnt.update(list(df2022["OrgSize"]))

org_cnt.most_common()
```
