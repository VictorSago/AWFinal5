```python
new_langs_by_c = {}

for row in df2022.itertuples():
    if type(row.Country) is str:
        new_langs_by_c.setdefault(str(row.Country), {
            "total": 0,
            "lang_counter": Counter()
        })
        langs = str(row.LanguageHaveWorkedWith).split(";")
        new_langs_by_c[row.Country]["total"] += 1
        new_langs_by_c[row.Country]["lang_counter"].update(langs)

new_langs_by_c
```

```python
sorted_langs_by_country = {k: new_langs_by_c[k] for k in sorted(new_langs_by_c.keys())}

sorted_langs_by_country
```

```python
for country in sorted_langs_by_country:
    print("-" * 10)
    total = sorted_langs_by_country[country]["total"]
    print(f"{country} (total: {total}): ")
    for l, c in sorted_langs_by_country[country]["lang_counter"].most_common(10):
        print(f"\t{l}: {round(c / total * 100, 2)}%")
```
