```python
import pandas as pd
from collections import defaultdict, Counter

df2022 = pd.read_csv("../rawdata/2022/survey_results_public.csv", dtype=str)
df2022.head()
```

```python
lang_by_dev = {}

for row in df2022.itertuples():
    if type(row.DevType) is str:
        dev_types = row.DevType.split(';')

        for dev_type in dev_types:
            lang_by_dev.setdefault(dev_type, {
                'total': 0,
                'language_counter': Counter()
            })

            languages = str(row.LanguageHaveWorkedWith).split(";")
            lang_by_dev[dev_type]['language_counter'].update(languages)
            lang_by_dev[dev_type]['total'] += 1

lang_by_dev
```

```python
for dev_type, info in lang_by_dev.items():
    print(dev_type)

    for language, value in info['language_counter'].most_common(5):
        language_pct = (value / info['total']) * 100
        language_pct = round(language_pct, 2)

        print(f'\t{language}: {language_pct}%')
```
