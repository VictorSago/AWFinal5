```python
import pandas as pd
import numpy as np

df = pd.read_csv('../rawdata/2022/survey_results_public.csv')

df
```

```python
from collections import Counter

pairs = Counter()

for row in df.itertuples():
    if type(row.LanguageHaveWorkedWith) is str:
        langs = row.LanguageHaveWorkedWith.split(';')
        for i, l1 in enumerate(langs):
            for l2 in langs[i:]:
                if l1 != l2:
                    ix = f'{l1} , {l2}'
                    pairs.update([ix])

pairs
```

```python
pairs.most_common()
```

```python
sorted(pairs)
```

```python
sorted_pairs = {k: pairs[k] for k in sorted(pairs.keys())}
sorted_pairs
```
