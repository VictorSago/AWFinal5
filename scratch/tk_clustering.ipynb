```python
import pandas as pd
import numpy as np

df = pd.read_csv('../rawdata/2022/survey_results_public.csv')

df
```

```python
from collections import Counter

pairs = Counter()

for row in df.itertuples():
    if type(row.LanguageHaveWorkedWith) is str:
        langs = row.LanguageHaveWorkedWith.split(';')
        for i, l1 in enumerate(langs):
            for l2 in langs[i:]:
                if l1 != l2:
                    ix = f'{l1} , {l2}'
                    pairs.update([ix])

pairs
```

```python
pairs.most_common()
```

```python
sorted_pairs = {k: pairs[k] for k in sorted(pairs.keys())}
sorted_pairs
```

```python
all_langs = []

all_pairs = Counter()

for row in df.itertuples():
    if type(row.LanguageHaveWorkedWith) is str:
        langs = row.LanguageHaveWorkedWith.split(';')
        for l1 in langs:
            if l1 not in all_langs:
                all_langs.append(l1)
            for l2 in langs:
                if l1 != l2:
                    ix = f'{l1} , {l2}'
                    all_pairs.update([ix])

print(sorted(all_langs))
```

```python
lang_mat = pd.DataFrame(columns=sorted(all_langs), index=sorted(all_langs))
lang_mat
```

```python
for lp in all_pairs:
    l1, l2 = lp.split(",")
    l1 = l1.strip()
    l2 = l2.strip()
    lang_mat[l1.strip()][l2.strip()] = all_pairs[f'{l1} , {l2}']
```
